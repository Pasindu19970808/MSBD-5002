<h2> Plotting Eulers equation on Complex plane to understand Fourier Transform (Self teaching) </h2>

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = list(map(lambda x: math.cos(2*np.pi*x),np.arange(0,1,0.001)))
y = list(map(lambda x: math.sin(2*np.pi*x),np.arange(0,1,0.001)))
fig,ax = plt.subplots(1,1,figsize = (5,5))
ax.plot(x,y)
```

```python
#plotting the movement along a unit circle by varying time and frequency
#Eulers equation of e^ix = cos(x) + isin(x)
#There is also a frequency factor which allows us to vary the amount moved along the unit circle for a fixed period of time
def plot_winding(f,t):
    #real part
    x = list(map(lambda x: math.cos(2*np.pi*f*x),np.arange(0,t,0.001)))
    #imaginary part
    y = list(map(lambda x: math.sin(2*np.pi*f*x),np.arange(0,t,0.001)))
    fig,ax = plt.subplots(1,1,figsize = (5,5))
    ax.plot(x,y)
```

```python
#At a frequency of 0.1Hz, for 1 second we only move through 1/10 of the circle
plot_winding(0.1,1)
```

```python
#At a frequency of 0.1Hz, for 10 second we only move through the whole circle
plot_winding(0.1,10)
```

```python
#At a frequency of 20Hz, for 10 second we only move through the whole circle 20 times. Hence the thick plot showing over lap
plot_winding(20,10)
```

```python
def double_side_exp(a):
    t_minus = np.arange(-10,0,0.01)
    t_plus = np.arange(0,10,0.01)
    result = list(map(lambda x: np.exp(a * x), t_minus))
    result.extend(list(map(lambda x: np.exp(-a * x), t_plus)))
    fig,ax = plt.subplots(1,1,figsize = (8,8))
    ax.plot(np.concatenate((t_minus,t_plus),0),result)
```

```python
double_side_exp(0.1)
```

```python
def fourier_double_side_exp(a):
    f = np.arange(-20,20,0.01)
    result = list(map(lambda x: (2*a)/(a**2 + (2*np.pi*x)**2),f))
    fig,ax = plt.subplots(1,1,figsize = (8,8))
    ax.plot(f,result)
```

```python
fourier_double_side_exp(0.1)
```
